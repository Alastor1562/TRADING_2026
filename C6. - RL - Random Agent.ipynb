```python
import gymnasium as gym
import numpy as np
```

```python
env = gym.make("CartPole-v1", render_mode="human")
# Render mode "human" is used to visualize the environment

st, info = env.reset()   # t = 0, reset the environment and get the initial state

done = False
epoch_reward = 0  # De todas las decisiones tomadas en el episodio, esta variable acumula la recompensa total obtenida
total_steps = 0


# [0.28252081 0.55222311 0.62611525 0.38420371]
weights = np.random.rand(4)  # Random weights for a simple linear policy

def get_action(st):
    # A simple linear policy that decides action based on the state and random weights
    action = 1 if np.dot(weights, st) > 0 else 0  # Simple linear policy based on the state
    return action

while not done:
    action = get_action(st)  # Get action based on current state
    sp, reward, terminated, truncated, info = env.step(action)  # Take action in the environment
    epoch_reward += reward  # Accumulate reward
    total_steps += 1  # Increment step count
    done = terminated or truncated  # Check if episode is done
    st = sp

env.close()  # Close the environment after the episode is done
```

```python
print(f"Total reward: {epoch_reward}, Total steps: {total_steps}, Weights: {weights}")
```
